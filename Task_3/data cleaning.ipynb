1. **Invoice No**: Identifies unique transactions. Could help in understanding the number of purchases and frequency.    


2. **StockCode**: Identifies each product uniquely, useful for analyzing product performance or popularity.   


3. **Description**: Product names or details that could help in categorizing and spotting trends in what’s being bought.    


4. **Quantity**: How many units were sold per transaction. Useful for understanding high-volume products and customer purchasing behavior.    


5. **Invoice Date**: Helps analyze trends over time—like seasonal spikes or monthly performance.    


6. **Unit Price**: Allows you to calculate revenue per item and see which products contribute most to revenue.    


7. **Customer ID**: Lets you identify repeat purchases, analyze customer loyalty, and measure retention.    


8. **Country**: Helpful for geographic insights—like which regions are driving the most revenue.
